```python
# !pip install pandas scikit-learn

import tkinter as tk
from tkinter import ttk
import pandas as pd
import random
from sklearn.tree import DecisionTreeClassifier

# -----------------------------
# STATES & ZONES
# -----------------------------
state_zones = {
    "Rajasthan": "Dry", "Gujarat": "Dry",
    "Assam": "Rainy", "Meghalaya": "Rainy", "Arunachal Pradesh": "Rainy",
    "Tripura": "Rainy", "Nagaland": "Rainy", "Manipur": "Rainy", "Mizoram": "Rainy",
    "Himachal Pradesh": "Cold", "Uttarakhand": "Cold", "Ladakh": "Cold",
    "Jammu and Kashmir": "Cold", "Sikkim": "Cold",
    "Kerala": "Tropical", "Tamil Nadu": "Tropical", "Karnataka": "Tropical",
    "Andhra Pradesh": "Tropical", "Goa": "Tropical",
    "Madhya Pradesh": "Moderate", "Uttar Pradesh": "Moderate",
    "Bihar": "Moderate", "Jharkhand": "Moderate",
    "Punjab": "Moderate", "Haryana": "Moderate",
    "Chhattisgarh": "Moderate", "Odisha": "Moderate",
    "West Bengal": "Moderate", "Maharashtra": "Moderate",
    "Telangana": "Moderate",
    "Delhi": "Moderate", "Chandigarh": "Moderate",
    "Puducherry": "Tropical", "Lakshadweep": "Tropical",
    "Andaman and Nicobar Islands": "Rainy",
    "Dadra and Nagar Haveli and Daman and Diu": "Tropical"
}

# -----------------------------
# ML MODEL
# -----------------------------
data = {
    'Temperature': [35, 30, 20, 25, 28, 18, 32, 27, 22, 26],
    'Humidity':    [30, 60, 70, 80, 65, 50, 75, 68, 55, 72],
    'Rainfall':    [50, 200, 300, 250, 150, 100, 220, 180, 120, 260],
    'Crop': ['Bajra', 'Rice', 'Tea', 'Sugarcane', 'Maize',
             'Wheat', 'Rice', 'Cotton', 'Gram', 'Jute']
}

df = pd.DataFrame(data)
X = df[['Temperature', 'Humidity', 'Rainfall']]
y = df['Crop']

model = DecisionTreeClassifier()
model.fit(X, y)

# -----------------------------
# WEATHER
# -----------------------------
def generate_weather(zone, month):
    if zone == "Dry":
        temp = (30, 45); humidity = random.randint(20, 40); rainfall = random.randint(0, 100)
    elif zone == "Rainy":
        temp = (20, 32); humidity = random.randint(70, 95); rainfall = random.randint(200, 400)
    elif zone == "Cold":
        temp = (-5, 20); humidity = random.randint(40, 70); rainfall = random.randint(50, 150)
    elif zone == "Tropical":
        temp = (25, 35); humidity = random.randint(60, 90); rainfall = random.randint(150, 300)
    else:
        temp = (20, 35); humidity = random.randint(40, 70); rainfall = random.randint(50, 200)

    wind = random.randint(5, 30)
    aqi = random.randint(40, 200)

    if month in ["December","January","February"]: season="Winter"
    elif month in ["March","April","May"]: season="Summer"
    elif month in ["June","July","August","September"]: season="Monsoon"
    else: season="Autumn"

    return temp, humidity, rainfall, wind, aqi, season

# -----------------------------
# COLORS
# -----------------------------
BG = "#0f172a"
CARD = "#1e293b"
TEXT = "#e2e8f0"
ACCENT = "#38bdf8"

# -----------------------------
# PREDICT FUNCTION
# -----------------------------
def predict():
    state = state_var.get()
    month = month_var.get()

    if not state or not month:
        clear_output()
        error_label.config(text="⚠ Please select both fields")
        return

    error_label.config(text="")

    zone = state_zones[state]
    temp, humidity, rainfall, wind, aqi, season = generate_weather(zone, month)

    avg_temp = sum(temp)//2
    best_crop = model.predict([[avg_temp, humidity, rainfall]])[0]

    extra = list(set(y))
    extra.remove(best_crop)
    other_crops = random.sample(extra, 3)

    clear_output()

    # WEATHER OUTPUT
    add_row("📍 State:", state)
    add_row("🌤 Season:", season)
    add_row("🌡 Temperature:", f"{temp[0]}°C - {temp[1]}°C")
    add_row("💧 Humidity:", f"{humidity}%")
    add_row("🌧 Rainfall:", f"{rainfall} mm")
    add_row("🌬 Wind Speed:", f"{wind} km/h")
    add_row("🌫 AQI:", aqi)

    # CROPS
    tk.Label(output_frame, text="\n🌱 Recommended Crops:",
             font=("Segoe UI", 13, "bold"),
             fg=ACCENT, bg=CARD).pack(anchor="w")

    tk.Label(output_frame,
             text=f"⭐ Best Crop: {best_crop}",
             font=("Segoe UI", 12, "bold"),
             fg="#FFD700",
             bg=CARD).pack(anchor="w", pady=(5,5))

    tk.Label(output_frame,
             text="Priority List:",
             font=("Segoe UI", 11, "bold"),
             fg=TEXT, bg=CARD).pack(anchor="w")

    colors = ["#22c55e", "#facc15", "#fb923c"]

    for i, crop in enumerate(other_crops):
        tk.Label(output_frame,
                 text=f"{i+1}. {crop}",
                 font=("Segoe UI", 11, "bold"),
                 fg=colors[i],
                 bg=CARD).pack(anchor="w")

# -----------------------------
# HELPERS
# -----------------------------
def add_row(label, value):
    frame = tk.Frame(output_frame, bg=CARD)
    frame.pack(anchor="w")

    tk.Label(frame, text=label,
             font=("Segoe UI", 10, "bold"),
             fg=TEXT, bg=CARD).pack(side="left")

    tk.Label(frame, text=f" {value}",
             font=("Segoe UI", 10),
             fg=TEXT, bg=CARD).pack(side="left")

def clear_output():
    for widget in output_frame.winfo_children():
        widget.destroy()

# -----------------------------
# GUI
# -----------------------------
root = tk.Tk()
root.title("AI Weather & Crop System")
root.geometry("600x650")
root.configure(bg=BG)

# TITLE
tk.Label(root, text="🌾 Smart Weather and Crops Recommendation Analyser",
         font=("Segoe UI", 20, "bold"),
         fg="#ff00ff",
         bg=BG).pack(pady=12)

# INPUT
input_frame = tk.Frame(root, bg=CARD, padx=20, pady=15)
input_frame.pack(pady=10, fill="x", padx=20)

tk.Label(input_frame, text="State", font=("Segoe UI", 10, "bold"),
         fg=TEXT, bg=CARD).pack(anchor="w")

state_var = tk.StringVar()
ttk.Combobox(input_frame, textvariable=state_var,
             values=sorted(state_zones.keys())).pack(fill="x")

tk.Label(input_frame, text="Month", font=("Segoe UI", 10, "bold"),
         fg=TEXT, bg=CARD).pack(anchor="w", pady=(10,0))

month_var = tk.StringVar()
ttk.Combobox(input_frame, textvariable=month_var,
             values=["January","February","March","April","May","June",
                     "July","August","September","October","November","December"]
            ).pack(fill="x")

tk.Button(root, text="Predict", command=predict,
          bg=ACCENT, fg="black",
          font=("Segoe UI", 11, "bold")).pack(pady=10)

error_label = tk.Label(root, text="", fg="red", bg=BG)
error_label.pack()

# -----------------------------
# SCROLLABLE OUTPUT
# -----------------------------
container = tk.Frame(root, bg=BG)
container.pack(pady=10, fill="both", expand=True, padx=20)

canvas = tk.Canvas(container, bg=BG, highlightthickness=0)
scrollbar = ttk.Scrollbar(container, orient="vertical", command=canvas.yview)

scrollable_frame = tk.Frame(canvas, bg=CARD, padx=20, pady=15)

scrollable_frame.bind(
    "<Configure>",
    lambda e: canvas.configure(scrollregion=canvas.bbox("all"))
)

canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")

canvas.configure(yscrollcommand=scrollbar.set)

canvas.pack(side="left", fill="both", expand=True)
scrollbar.pack(side="right", fill="y")

output_frame = scrollable_frame

# Mouse scroll
def _on_mousewheel(event):
    canvas.yview_scroll(int(-1*(event.delta/120)), "units")

canvas.bind_all("<MouseWheel>", _on_mousewheel)

root.mainloop()
```
